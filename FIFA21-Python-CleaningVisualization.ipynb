Dataframe with only the important data

```python
df2 = df[['Name', 'Age', 'Team & Contract', 'Value', 'Wage', 'Joined', 'Release Clause', 'Height', 'Weight', 'foot']]
df2
```

Deleting the \n in Team & Contract column

```python
# checking the datatype
column_data_type = df2['Team & Contract'].dtype
print(column_data_type)
```

```python
df2['Team & Contract'] = df['Team & Contract'].str.replace('\n','')
```

```python
df2['Team & Contract']
```

Splitting the Team & Contract column

```python
# defining the column names for the split data
column_names = ['Club','Contract until']
```

```python
# defining the column names for the split data
column_names = ['Club','Contract until']

# split the data into "Club" and "Contract until"
splitdata = df2['Team & Contract'].str.split('~', expand = True)
splitdata.columns = column_names
splitdata
```

```python
# split the "Joined" date still located in "Club" column into a separate column
splitdata['Joined'] = splitdata['Club'].str.extract(r'(\d{4})')
# erase the Year from Club column
splitdata['Club'] = splitdata['Club'].str.replace(r'\d{4}', '').str.strip()

  ## r   > indicates raw strings
  ## \n  > only DIGITS (inside characters)
  ## {4} > extracts 4 consecutive digits

## if one player is on loan then the "loaned until" date and on loan text will also bi present in the club name
## we will leave that as it is because those players wont be regarded in the further visualisations
```

```python
splitdata
```

Appending the new columns into the dataframe

```python
# we already have the Joined date in the dataframe so no need to append it into the table
# append the club and contract until
df2 = df2.assign(Club = splitdata['Club'], Contract_Until = splitdata['Contract until'])
df2
```

Split the "Joined" column into Year, Month, Day columns

```python
df2['Joined']
```

```python
column_data_type = df2['Joined'].dtype
print(column_data_type)
```

```python
# split the data
date_parts = df2['Joined'].str.split(' ', expand = True)
df2['Year_Joined'] = date_parts[2]
df2['Month_Joined'] = date_parts[0]
df2['Day_Joined'] = date_parts[1].str.rstrip(',')
```

```python
df2
```

Value/Wage/Release Clause into Numeric values

```python
value_columns = df2[['Value', 'Wage', 'Release Clause']]
print(value_columns)
```

```python
# erase the € sign
columns_to_convert = ['Value', 'Wage', 'Release Clause']
for col in columns_to_convert:
  df[col] = df[col].str.replace('€', '')
```

```python
columns_to_convert = ['Value', 'Wage', 'Release Clause']
for col in columns_to_convert:
    df[col] = df[col].str.replace('M', '00000')
    df[col] = df[col].str.replace('K', '000')
    df[col] = df[col].str.replace('.', '')
```

```python
df['Value'] = df['Value'].astype('int')
df['Wage'] = df['Wage'].astype('int')
df['Release Clause'] = df['Release Clause'].astype('int')
```

```python
df['Value']
```

```python
# Append the corrected columns
df2 = df2.assign(Value_INT = df['Value'], Wage_INT = df['Wage'], Release_Clause_INT = df['Release Clause'])
df2
```

Height & Weight into cms & kgs

```python
two_columns = df[['Height', 'Weight']]
two_columns
```

```python
# Height
def height_to_cm(height):
    feet, inches = height.split("'")
    inches = inches.replace('"', "")  # Remove the quotation mark from inches
    cm = int(feet) * 30.48 + int(inches) * 2.54
    return cm

# Convert the "Height" column from feet'inches to centimeters
two_columns['Height_CM'] = two_columns['Height'].apply(height_to_cm)
```

```python
two_columns['Height_CM'] = two_columns['Height_CM'].astype('int')
```

```python
two_columns['Height_CM']
```

```python
# Weight
two_columns['Weight'] = two_columns['Weight'].str.replace('lbs', '')
```

```python
weight_data_type = two_columns['Weight'].dtype
print(weight_data_type)
```

```python
two_columns['Weight'] = two_columns['Weight'].astype('int')
```

```python
lbs_to_kg = 0.45359237
two_columns['Weight_KG'] = two_columns['Weight'] * lbs_to_kg
two_columns['Weight_KG'] = two_columns['Weight_KG'].astype('int')
```

```python
two_columns['Weight_KG']
```

```python
# Append the two new columns
df2 = df2.assign(Height_CM = two_columns['Height_CM'], Weight_KG = two_columns['Weight_KG'])
df2
```

Final dataset

```python
df_final = df2[['Name', 'Age', 'Club', 'Contract_Until', 'Year_Joined', 'Month_Joined', 'Day_Joined', 'Value_INT', 'Wage_INT', 'Release_Clause_INT', 'Height_CM', 'Weight_KG']]
df_final
```

Visualisations

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Value vs. Wage
# Which players are highly valuable but still underpaid (on low wages)?
# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df_final['Wage_INT'], df_final['Value_INT'], alpha=0.7)
plt.title('Value vs. Wage')
plt.xlabel('Wage')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```

```python
# Top 10 clubs per AVG Wage
# Filter out rows with clubs containing "On Loan"
# Bar chart
df_final = df_final[~df_final['Club'].str.contains("On Loan")]

# AVG wage for each club
club_wages = df_final.groupby('Club')[['Wage_INT']].mean()

# Sort clubs by AVG wage
club_wages = club_wages.sort_values(by='Wage_INT', ascending=False)

# Select the top 10 clubs
top_wages = club_wages.head(10)

plt.figure(figsize=(12, 6))
top_wages.plot(kind='bar', stacked=False, legend=False)
plt.title('Top 10 Clubs based on average Wages')
plt.xlabel('Club')
plt.ylabel('Amount')
plt.xticks(rotation=90)
plt.show()
```

```python
# Value vs age
# Is there any correlation between the age of players and their market value?
# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df_final['Age'], df_final['Value_INT'], alpha=0.5)
plt.title('Value vs. Age')
plt.xlabel('Age')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```

```python
# Age distribution
# Question/Topic: How is the age distribution of soccer players in the dataset?
# Histogram
plt.figure(figsize=(10, 6))
plt.hist(df_final['Age'], bins=20, edgecolor='black', alpha=0.7)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Age distribution
# Kernel density plot

plt.figure(figsize=(10, 6))
sns.kdeplot(df_final['Age'], fill=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
# Age distribution
# Violin plot
club_release_clauses = df_final.groupby('Club')['Release_Clause_INT'].mean()

# Sort clubs by average release clause
club_release_clauses = club_release_clauses.sort_values(ascending=False)

# Select the top 10 clubs
top_clubs = club_release_clauses.head(10)

# Filter the dataset to include only rows for the top clubs
df_top_clubs = df_final[df_final['Club'].isin(top_clubs.index)]

plt.figure(figsize=(10, 6))
plt.xticks(rotation=45)
sns.violinplot(data=df_top_clubs, x='Club', y='Release_Clause_INT')
plt.title('Top 10 Clubs based on Release Clauses')
plt.xlabel('Club')
plt.ylabel('Release Clause')
plt.tight_layout()
plt.show()
```
